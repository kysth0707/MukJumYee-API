```python
VIDEO_RESOLUTION = (9*20, 19*20)
```

```python
cap = cv.VideoCapture(r'E:\GithubProjects\HIDDEN\ImageFromVideo\tests\test.mp4')

images = []
for _ in range(6):
	ret, frame = cap.read()
	if not ret:
		break
	frame = cv.flip(cv.flip(frame, 0), 1)
	frame = cv.resize(frame, VIDEO_RESOLUTION)
	images.append(frame)
	if cv.waitKey(1) & 0xFF == ord('q'):
		break
images = np.array(images)
cap.release()
cv.destroyAllWindows()
```

왜곡이 약 6부터 줄어드는 듯

```python
stitcher = cv.Stitcher_create()

retval, stitchedImage = stitcher.stitch(images[0:6])
plt.imshow(stitchedImage)
```
