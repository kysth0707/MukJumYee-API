```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
VIDEO_RESOLUTION = (9*20, 19*20)
```

```python
cap = cv.VideoCapture('test.mp4')

ret, frame = cap.read()
frame = cv.flip(cv.flip(frame, 0), 1)
frame = cv.resize(frame, VIDEO_RESOLUTION)
frameA = frame

for _ in range(5):
    ret, frame = cap.read()

ret, frame = cap.read()
frame = cv.flip(cv.flip(frame, 0), 1)
frame = cv.resize(frame, VIDEO_RESOLUTION)
frameB = frame

	
cap.release()
```

```python
plt.imshow(frameA)
```

```python
plt.imshow(frameB)
```
